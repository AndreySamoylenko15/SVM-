```python
import pandas as pd
import numpy as np
import requests
from matplotlib import pyplot as plt
import gdown
import os
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import zipfile
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
def check_csv_structure(data_dir):
    structure = {}
    for root, dirs, files in os.walk(data_dir):
        for file in files:
            if file.endswith(".csv"):
                file_path = os.path.join(root, file)
                try:
                    df = pd.read_csv(file_path, nrows=1)  # Зчитуємо перший рядок
                    structure[file] = df.columns.tolist()
                except Exception as e:
                    print(f"Помилка при зчитуванні {file}: {e}")
    return structure

data_dir = "data"
structure = check_csv_structure(data_dir)

for file, columns in structure.items():
    print(f"Файл: {file} -> Колонки: {columns}")
```

```python
def load_all_csv_files_with_labels(data_dir):
    all_files = []
    all_labels = []

    for root, dirs, files in os.walk(data_dir):
        for file in files:
            if file.endswith(".csv"):
                file_path = os.path.join(root, file)

                # Визначаємо мітку на основі назви папки
                label = os.path.basename(os.path.dirname(file_path))

                # Завантажуємо дані
                df = pd.read_csv(file_path)
                df['label'] = label  # Додаємо колонку з міткою

                all_files.append(df)

    # Об’єднуємо всі DataFrame в один
    combined_df = pd.concat(all_files, ignore_index=True)
    return combined_df

# Завантажуємо всі CSV-файли з папки "data" з мітками
data_dir = "data"
df = load_all_csv_files_with_labels(data_dir)

print("Об'єднаний датасет:", df.shape)
print(df.head())
print("Унікальні мітки:", df['label'].unique())
```

```python
def calculate_time_features(df):
    # Додаємо нові колонки безпосередньо до основного DataFrame
    df['mean'] = df[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z']].mean(axis=1)
    df['std'] = df[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z']].std(axis=1)
    df['max'] = df[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z']].max(axis=1)
    df['min'] = df[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z']].min(axis=1)
    df['median'] = df[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z']].median(axis=1)
    return df

# Оновлюємо основний DataFrame
df = calculate_time_features(df)

print("Колонки після додавання часових ознак:")
print(df.columns)
print(df.head())

```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['mean', 'std', 'max', 'min', 'median']])
df[['mean', 'std', 'max', 'min', 'median']] = scaled_features
```

```python
# Вибираємо ознаки (features) та мітки (labels)
X = df[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z', 'mean', 'std', 'max', 'min', 'median']]
y = df['label']

# Масштабування ознак
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Розділяємо на навчальні та тестові дані (80% навчання, 20% тест)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("Розмірність X_train:", X_train.shape)
print("Розмірність y_train:", y_train.shape)
print("Розмірність X_test:", X_test.shape)
print("Розмірність y_test:", y_test.shape)
```

```python
# Створення та навчання моделі
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Прогнозування на тестових даних
y_pred = rf.predict(X_test)

# Оцінка моделі
accuracy = accuracy_score(y_test, y_pred)
print("Точність моделі Random Forest:", accuracy)
print("Звіт про класифікацію:")
print(classification_report(y_test, y_pred))
```

```python
scores = cross_val_score(rf, X_scaled, y, cv=5)
print("Крос-валідація (5 фолдів) - середня точність:", scores.mean())
print("Окремі фолди:", scores)

```
